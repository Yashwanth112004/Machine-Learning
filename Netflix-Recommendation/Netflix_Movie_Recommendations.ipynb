```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_set = pd.read_excel('/content/sample_data/Netflix_Recommandation.xlsx', header=None)
transactions = []
for i in range(0, 7465):
    transactions.append([str(data_set.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
basket_intelligence = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(basket_intelligence)
results
```

```python
def inspect(results):
    movie_1 = [tuple(result[2][0][0])[0] for result in results]
    movie_2 = [tuple(result[2][0][1])[0] for result in results]
    support = [result[1] for result in results]
    confidence = [result[2][0][2] for result in results]
    lift = [result[2][0][3] for result in results]
    return list(zip(movie_1, movie_2, support))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Movie 1', 'Movie 2', 'Support'])
```
